### Polymorphism. 
Polymorphism is a core concept in object-oriented programming(OOP), that allows objects of different classes to be treated as objects of common superclass. It provides a way to perform a single action in different forms. Polymorphism is typically achieved through method overriding and interfaces.

### Method Overriding 
Method Overriding allows a child class to provide a specific implementation of a method that is already defined in its parent class.

```python
##base class
class Animal:
    def speak(self):
        return "sound of the animal" 

##derived class 1
class Dog(Animal):
    def speak(self):
        return "woof"  ## this method is already present in parent class above but the child class has provided its own specific implementation 

##derived class 1
class Cat(Animal):
    def speak(self):
        return "meawo"  ##overwriting the parent class  method by specific method in child class



## function that demonstarate polymorphisum
"""it allows objects of different classes to be treated 
through the same interface which is animal_speak() in this and animal1 
is the object which can be from any class Dog or Cat"""
def animal_speak(animal1):
    print(animal1.speak())   ##animal1 will be an object 


dog1=Dog()
cat1=Cat()
print(dog1.speak())
print(cat1.speak())
animal_speak(dog1)  ##SINCE dog1 is from dog class it will call its specific method
                    ## animal1 ma dog1 jani vayo ni ta
```

```python
##polymorphisum with function and methods


##base class
class shape:
    def area(self):
        print("the area is")

##derived class 1
class rectangle(shape):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth

    def area(self):
        return self.length*self.breadth
    
class circle(shape):
    def __init__(self,radius):
        self.radius=radius
    

    def area(self):
        return 3.14*self.radius*self.radius
    
##polymorphisum
def print_area(shape1):
    print("the area is ",shape1.area() )

rect1=rectangle(2,3)
circ1=circle(3)
print_area(rect1)
print_area(circ1)

```
