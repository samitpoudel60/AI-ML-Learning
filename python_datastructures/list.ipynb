slicing lists


```python
##list slicing
numbers = [10, 20, 30, 40, 50, 60, 70, 80, 90]
print("Original numbers list:", numbers)
```

```python
print(numbers[2:5])##slicing from index 2 to 4
print(numbers[:4])##slicing from start to index 3
print(numbers[5:])##slicing from index 5 to end
print(numbers[::2])##slicing with step 2
print(numbers[::-1])##reversing the list using slicing
```

```python
numbers[::3] 
```

```python
##iterating
for number in numbers:
    print(number)
```

```python
##iterationn with index (enumerate() function)
for index, number in enumerate(numbers):
    print(index,number)
```

```python
##list comprehension
list=[]
for x in range(10):
    list.append(x**2)

print(list)
```

```python
##this can be written in another simple way as
##[operation , iteration]

[x**2 for x in range(10)]
##in this way it already forms a list
```

##list comprehension

basic syntax ::

[expression/operation for item in iteration]
____________________________________________________
with condition logic ::

[expression/operation for item in iteration if condition ]
______________________________________________________
nested list comprehension ::

[expression/operation for item1 in iteration1 for item2 in iteration2]

```python
##bassic list comprehension
square=[num**2 for num in range(10)]
print(square)
```

```python
##list comprehension with condition
list=[]
for i in range(10):
    if i%2==0:
        list.append(i)

print(list)
```

```python
#3now this can be written as
even_numbers = [num for num in range(10) if num % 2 == 0]
print(even_numbers)
```

```python
##nested list comprehension

list1=[1,2,3]
list2=['a','b','c','d']

pair=[(x,y) for x in list1 for y in list2]
print(pair)
```

```python
##another a*b multiply example
a=[1,2,3]
b=[4,5,6]
product=[i*j for i in a for j in b]
print(product)
```

```python
##list comprehension with function calls
words=["hello","world","python","programming"]
lengths=[len(word) for word in words]
print(lengths)##output: [5, 5, 6, 11]
```
