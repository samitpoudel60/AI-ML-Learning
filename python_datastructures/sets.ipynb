Sets are built-in data types in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entities, and performing mathematical set operations like Union, Intersection, Difference, and Symmetric Difference.

```python
##creating a set
my_set={1,2,3,4}
print(my_set)
print(type(my_set))
```

```python
set1=set([1,2,2,3,3,4,5])##duplicate elemets are not included
print(set1)
```

```python
##basic set operation
my_set1={1,2,3,4,5,6}
my_set1.add(6)
print(my_set1)##duplicate so does not add
my_set1.add(7)
print(my_set1)
```

```python
##removing the elemets from the set
my_set1.remove(1)
print(my_set1)

```

```python
#set.remove(10)  ##10 is not in the set so it gives error
```

```python
my_set1.discard(10)##it check and remove an item if it is present in the set
print(my_set1)
```

```python
##pop method
my_set={1,2,3,4,5}
remove_element=my_set.pop()
print(remove_element)
##it uses fifo principle first in first out
```

```python
##cleasr all the elemnts
my_set.clear()
print(my_set)
```

```python
#set membership tst
my_set1={1,2,3,4,5,6,7}
print(3 in my_set1)
print(10 in my_set1)

```

```python
##mathematical operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

union_set=set1.union(set2)
print("uinion set is",union_set)

interserction_set=set1.intersection(set2)
print("intersection is",interserction_set)

set1.intersection_update(set2)
print(set1)##the intersection is done and ans is upadted in the set 1

```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

#difference
differenc=set1.difference(set2)##yo bahneko (set1-set2) ho
print(differenc)
##we also have difference_update like before
```

```python
#sets methhods
set3={1,2,3,4,5}
set4={2,3,4}
#is subset
print(set3.issubset(set4))
print(set4.issubset(set3))
```

```python
##counting unique words in the text
text="my name is samit poudel i am engineering student"
words=text.split()
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
