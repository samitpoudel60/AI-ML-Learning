tuples are ordered collections of item which are immutable 

they are similar to lists but cannot be changed once created

tuples are defined using parentheses ()

```python
##creating a tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()##this is an emty list
print(type(lst))
tpl=tuple()##this is an empty tuple
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5])##changing a list into tuple
print(numbers)
```

```python
mixed_tuple=(1,"hello",3.4,True)
print(mixed_tuple)
```

```python
print(numbers[0])##first element
print(numbers[2])##third element
print(numbers[-1])##last element
print(numbers[1:4])##slicing
##tuples are immutable
#numbers[0]=10          ##this will raise an error
```

```python
num=(1,1,2,3,4,5)
print(num)
```

```python
print(num.count(1))##counting occurence of 1 in the tuple
print(num.index(3))##finding index of element 3 in the tuple
```

```python
##packing and unpacking tuple
packed_tuple="samit",25,"engineer"##yesari define garyo vane ni tuple nai hunxa
print(packed_tuple)
##unpacking a tuple
name,age,profession=packed_tuple
print(name)
print(age)
print(profession)

```

```python
##unpacking with stars
number_tuple=(1,2,3,4,5,6,7,8,9)
first,*middle,last=number_tuple
print(first)##what * does it it take all remaining elements in the middle
print(middle)
print(last)
```

```python
##nested list

nested_list=[[1,2,3],[4,5,6],[7,8,9]]
nested_list[0][1]##accessing index 1 from list0
nested_list[2][0:3]##accessing all elements from list2
```

```python
##nested tuples
nested_tuple=((1,2,3),("a","b","c"),(True,False))
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
##iterating over nested tuple
for tuple in nested_tuple:
    for item in tuple:
        print(item,end=" ")##end is used for printing in the same line 
                          ##even after each loop every thing print on same line
    print( )#this prints a blank line 
    ##it is necessary as it moves the cursor in the new line
```
