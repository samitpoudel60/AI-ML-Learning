```python
##data serialization
import json
data={"name":"samit", "age":25}
print(type(data))
#now converting this dictionary in json we will use
json_str=json.dumps(data)
print(json_str)
print(type(json_str))
```

```python
##CONVERTING IT BACK INTO THE DICT
json_parsed=json.loads(json_str)
print(json_parsed)
print(type(json_parsed))
"""whenever we get json from an api 
it will be in the form of this string format 
and in this way we can conver it into the dict 
and perform data sequentialization"""
```

```python
##csv library
import csv
with open("example.csv",mode="w",newline="") as file:
    writer=csv.writer(file)##this is to write in the file this is a writer function
    writer.writerow(["name","age"])
    writer.writerow(["samit",22])

with open("example.csv", mode="r") as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)

```

```python
##datetime library
from datetime import datetime,timedelta

now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)
##timedelta go 1 day back and calculate the time
print(yesterday)
```
