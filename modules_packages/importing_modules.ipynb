### importing modules in python:modules and packages

in python modules and packages help organize and reuse code> here is a comprehensive guide on how to import them.

```python
import math
print(math.sqrt(16))
```

```python
from math import sqrt, pi
##now that i have imported math i can simply write
print(sqrt(16))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4,5])
```

```python
from math import *
## * means it is importing all the functions of he math i dont have to particularly import some function
print(sqrt(16))
print(pi)
```

```python
##importing from a custom package we made ourself
from custom_package import math1

print(math1.sub(2,4))
```

```python
##this can also be importedd as
from custom_package.math1 import sub,add

print(add(11,8))
```

```python
from custom_package.math1 import add

print(add(11,8))
```

```python
from custom_package import math1

print(math1.sub(2,3))
```

```python
from custom_package import math1

print(math1.add(2,7))
```

```python
from custom_package.math1 import *

print(sub(3,2))
print(add(3,7))
```
