### Magic Methods
Magic Methods in Python also known as Dunder Methods( Double Underscore Methods )are special methods that start and end with double underscore. These methods enable you to define the behavior of objects for built-in operations such as arithmetic operations, comparisons, and more. 
### Magic Methods 
Magic Methods are predefined methods in Python that you can override to change the behavior of your objects. Some common Magic Methods include 

init-- Initialize a new instance of a class, 

str--Returns a string representation of an object, 

repr--Returns an official string representation of an object.

```python

##basic methods
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

person1=Person("samit",33)
print(person1)
    
##by default i will get this kind of output and let say i have to overwrite this so in thaT case i have to overwrite the __str__
```

```python
## another way can be using repr
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def __str__(self):
        return f"{self.name},{self.age} years old"
    
    def __repr__(self):
        return f"Person(name={self.name},age={self.age})"

person1=Person("samit",33)
print(person1)
print(repr(person1))
```

### OPERRATOR OVERLOADING


What is Overriding in __str__ and __repr__?

In Python, overriding __str__ and __repr__ means redefining these magic methods in your class to replace the default implementations provided by Python‚Äôs base object class.

By default, every class inherits from object, which already has __str__ and __repr__.

Default Behavior (Without Overriding)

    pass

‚úî To provide a human-readable description

‚úî Helps in debugging and logging

```python
"""
What is Operator Overriding?

Operator overriding (also called operator overloading) is the ability to change the behavior of operators (+, -, ==, <, etc.) for user-defined objects by overriding magic methods (dunder methods) such as __add__, __eq__, __lt__, etc.

In Python, operators do not work on objects by default unless we define how they should behave.

Why is Operator Overriding Necessary?
1Ô∏è‚É£ Default operators don‚Äôt work for custom objects
class Point:
    def __init__(self, x):
        self.x = x

p1 = Point(2)
p2 = Point(3)

# p1 + p2   ‚ùå TypeError (without overriding)


Python doesn‚Äôt know how to add two Point objects.

2Ô∏è‚É£ To give meaning to operators for objects
class Point:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return Point(self.x + other.x)


Now:

p3 = p1 + p2


‚úî + now has a logical meaning for Point.


"""


```

```python

## mathematical operation of the vectors
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # addition (+)   ##other will be the another vector point
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # subtraction (-)
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    # multiplication (*)
    def __mul__(self, other):
        return Vector(self.x * other.x, self.y * other.y)

    def __eq__(self,other):
        return self.x==other.x and self.y==other.y  ##returns either true or false
    
    #just to represents the points
    def __str__(self): 
        return f"(points are :{self.x}, {self.y})"


## creating the objects of the vector class
v1=Vector(2,3)
v2=Vector(3,4)
print(v1+v2)
print(v1==v2)## returns true or false
```
