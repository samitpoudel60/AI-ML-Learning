```python
## best way to define attribute
class dog:
    ##constructor
    def __init__(self,name,age):##they help to assign the parameters at runtime
        self.name=name
        self.age=age ## these are the instance variables at runtime

##creating objects
dog1=dog("sam",7)  ##since the constructor has two parameters it must have two arguments
print(dog1)
print(dog1.name)
print(dog1.age)


## if i have to add more attribute like breed, color i must assign it in the constructor
```

```python
##define a class with instace method

class dog:
    def __init__(self,name,age):##they help to assign the parameters at runtime
        self.name=name
        self.age=age ## these are the instance variables at runtime
    
    def bark(self):
        print(self.name,"says woof")

        
##creating objects
dog1=dog("sam",7)  
dog1.bark()
dog2=dog("luffy",3)  
dog2.bark()
```

```python
## modelling a banlk account
## define a class for bank account

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,ammount):# i have to deposite (ammount)
        self.balance+=ammount  ##the the ammount will be added to the balance
        print(ammount,"is deposited.","new balance is",self.balance)

    def withdraw(self,ammount):
        if ammount>self.balance:
            print("you do not haver that much balance in your account")
        else:
            self.balance-=ammount
            print(ammount,"has been withdrawn","your balance is",self.balance)
    
    def get_balance(self):
        return self.balance

##CREATING OBJECT
user1=BankAccount("SAMIT POUDEL",5000)
print(user1.balance)
```

```python
#calling instance method
user1.deposit(1000)##since there is one parameter i have to enter an argument
```

```python
user1.withdraw(300)
```
