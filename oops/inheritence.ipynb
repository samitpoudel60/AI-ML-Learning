### Inheritance in Python. 
Inheritance is a fundamental concept in object-oriented programming, OOP, that allows a class to inherit attributes and methods from another class. This lesson covers single inheritance, multiple inheritance, demonstrating how to create and use them in Python.

```python
##inheritance (single inheritance)
##class inhertited from one base class
## Parent class

class Car:
    def __init__(self,windows,doors,engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type 

    def drive(self):
        print("the person is driving",self.engine_type,"car")

car1=Car(4,2,"mechanic")
car1.drive()


```

```python
## now supoose a new class have to be created tesla which parent class is the Car class
## we can inheritant the attributes and instance methods of the parent class


class Tesla(Car): ## is_selfdriving attribute is added in this
    def __init__(self,windows,doors,engine_type,is_selfdriving):
        super().__init__(windows,doors,engine_type)     #super()is used to inheritant the attributes
        self.is_selfdriving=is_selfdriving ## this is not available in the parent class


    def selfdriving(self):
        print("tesla supports self driving:",self.is_selfdriving)

tesla1=Tesla(3,2,"electric",True)
tesla1.selfdriving()
```

```python
##we can also call parent class functions like
tesla1.drive()
```

```python
##3 multiple inheritence
## when a class is inherits from multiple base class


##base class 1
class Animal:
    def __init__(self,name):
        self.name=name
    
    def speak(self):
        print("subclass must implement this method")


##base class 2
class Pet():
    def __init__(self,owner):
        self.owner=owner


##derived class
class Dog(Animal,Pet):
    def __init__(self,name,owner,age):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)#way of inheriting attributes in multiple inheritence
        self.age=age

    def speak(self):
        return( self.name,"says woof")
    
dog1=Dog("sam","samit",2)
dog1.speak()

##there are two speak but since we are using dog class it will give priority to its own speak()



    
```
