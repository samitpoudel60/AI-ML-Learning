### Abstarction
Abstraction is the concept of hiding complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

```python
from abc import ABC, abstractmethod  ## for abstraction we use abc nad import ABC- (abstract base class)

##abstract class
class Vehicle(ABC):
    def drive(self):
        print("the vehicle i s used for driving")

    @abstractmethod  ## this should be completely empty
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("car engine started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car1=Car()
operate_vehicle(car1)
```

Vehicle is an abstract class

start_engine() is an abstract method

Any class that inherits Vehicle must implement start_engine()

Car implements the abstract method

Now Car can be instantiated

❌ If start_engine() was missing → Python would throw an error:
