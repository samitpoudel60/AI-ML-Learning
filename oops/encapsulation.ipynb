### Encapsulation and Abstraction 
Encapsulation and abstraction are two fundamental principles of object-oriented programming that helps in designing robust, maintainable, and reusable code. Encapsulation involves bundling data and methods that operate on the data within a single unit, while abstraction involves hiding complex implementation details and exposing only necessary features.

Encapsulation is the concept of wrapping the data, variables, and method functions together as a single unit. It restricts direct access to some of the object components, which means preventing accidental interference and misuse of the data.

Public variables are accessible from anywhere inside or outside the class, it is the default behavior in the Python,

 whereas the protected variables are intended to be accessed within the class and its subclass, and 
 
 the private variables can be only accessed within the class.


```python
### encapsulation with getter and setter method
### public ,protected and private vaeriabls?access methods

class Person:
    def __init__(self,name,age):
        self.name=name ##public variables
        self.age=age


def get_name(person): ## accessed outside of the class
    return person.name

person1=Person("samit poudel",34)
print(get_name(person1))
```

protected variables say that we cannot access it from the outside of the class but can access it from the derived class


```python
class Person:
    def __init__(self,name,age,gender):
        self._name=name ##protected variables
        self._age=age   ## protected variables single underscore _
        self.gender=gender


class Employee(Person):
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)

employee1=Employee("samit",23,"male")
print(employee1._name)
## access garna ko lagi  _name, _age nai hunxa name, age hunna
```

```python
## encapsulation with getter and setter method
class Person:
    def __init__(self,name,age):
        self.__name=name ##private variables/access method
        self.__age=age  ## private variables 

    ##getter method for name
    def get_name(self):
        return self.__name

    ##setter method for name to set name or change name
    def set_name(self,name):
        self.__name=name

    ##getter method for age
    def get_age(self):
        return self.__age

    ##setter method for age to set or change age
    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print("age cannot be negative")

person3=Person("samit",13)

## access and modify private variables using getter and setter method

print(person3.get_name())
print(person3.get_age())

person3.set_age(35)
print(person3.get_age())

person3.set_age(-34)


    
```
