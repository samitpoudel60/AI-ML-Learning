### Custom exeption (Raise and Thow an exeption)


```python
class Error(Exception):## here we make a class Error and inherit the Exception
    pass

class dob_exception(Error):## here we make a class to handle the date of birth exeption and inherit the Error class
    pass
```

```python
## handeling exeption not letting people above 30 and below 20 years of age fill the forum
dob=int(input("enter your date of birth"))
age=2025-dob

try:
    if age>20 and age<30:
        print("you can fill the forum")
    else:
        raise dob_exception
    
except dob_exception:
    print("sorry your age is not valid to fill the forum")

    

    
    
```
