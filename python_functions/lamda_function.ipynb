
Lambda functions in Python.
 
Lambda functions are small, anonymous functions defined using the lambda keyword. They can have any number of arguments, but only one expression. They are commonly used for short operations or as arguments to higher-order function

lambda function syntax is given as:

{lambda parameter/argument: expression}

```python
## for example we have a simple function\
def add(a,b):  ##a and b are the argument
    return a+b  ## a+b is the expression

add(2,3)
```

```python
##we can write this in easier way bt using lambda function as

addition= lambda a,b:a+b ##addition is a lamda function
type(addition)
```

```python
add(2,3)
```

```python
def even(num):
    if num%2==0:
        return True
    
even(24)
```

```python
##using lamda function 
is_even= lambda n:n%2==0
is_even(12)
```

```python
def add(a,b,c):
    return a+b+c
add(1,2,3)
```

```python
#3using lambda
add1= lambda x,y,z:x+y+z
add1(12,13,14)
```


MAP FUNCTION

a map function is a function that applies a function to all the items in a list or collection

```python
##EXAMPLE 
numbers=[1,2,3,4,5,6,7]
for number in numbers:
    def square(number):
        return number**2
    
for number in numbers:
    print(square(number))
```

```python
## but using a map/lambda function we can do it in an easy way
num_list=[1,2,3,4,5,6,7]
list(map(lambda n:n**2,num_list))
```
