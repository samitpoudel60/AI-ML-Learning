```python
##now if i have to use this in multiple places then i will require a fucntion

def even_or_odd(num):
    #this fuction finds whether the number is odd or even
    if num%2==0:
        print("even")
    else:
        print("odd")
```

```python
##variable length arguments
##positional and keyword arguments
# *args is a kind of positional argument we use

def print_numbers(*args): # *args will have multiple parameters
    for number in args:
        print(number)
```

```python
##now as the *args reperesnt multiple variables or parameter i can give multiple arguments
## *args represents all this positional arguments

print_numbers(1,2,3,4,5,6,"samit",9.99)
```

```python
##keyword argument
# we write it by **kwargs
##here all the arguments will be in the form of key value pairs
# so we can access this like we access dictionaries 

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(key,":",value)

```

```python
print_details(name="samit", age=32, country="nepal")
##all the parameters are represented by **kwargs
```

```python
##by using both of them we can write as

def print_details1(*args,**kwargs):
    print("positional argument")
    for val in args:
        print(val)
    print("keyword argument")
    for key,value in kwargs.items():
        print(key,":",value)
```

```python
print_details1(1, 2, 3, "samit", name="suraj", age=22, country="nepal")
##it automatically identifies positional arguyment and key word arguments
##In positional argument, we have only values, while in keyword argument, we have key and values pair.
```

```python
##return statement
def multiply(a,b):
    return a*b

multiply(2,4)
```

```python
#we can also return multiple parameters like

def multiply(a,b):
    return a,b,a*b

multiply(2,4)
```
