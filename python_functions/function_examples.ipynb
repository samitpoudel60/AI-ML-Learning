Example 1. Temperature Conversion

```python
def convert_temp(temp,unit):
    if unit=="C":
        return temp*9/5+32 ##converting celcius into ferhenheit
    elif unit=="F":
        return (temp-32)*5/9 ##converting ferheheit into celsius
    else:
        return None

print(convert_temp(25,"C"))

print(convert_temp(77,"F"))
    
```

Example 2. Password Strength Checker

```python
def is_strong_password(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in "!@#$%^&()_" for char in password):
        return False
    return True

##calling a function
print(is_strong_password("Wersr334"))

print(is_strong_password("Wersr33&4"))


```

Example 3. Calculate the total cost of items in a shopping cart 

```python
def calculate_total(cart):
    total_cost=0
    for item in cart:
        total_cost+= item["price"] * item["quantity"]

    return total_cost

cart=[
    {"name":"apple","price":0.5,"quantity":5},
    {"name": "banana", "price": 0.73, "quantity": 12},
    {"name": "milk", "price": 1.2, "quantity": 2}
]
##calling the function
print(calculate_total(cart))
```

Example 4. Check if a string is palindrome


```python
def is_palindrome(string):
    string=string.lower().replace(" ","")##replace empty space with ""
    print(string[::-1])
    return string==string[::-1]## string[::-1] is reverse of string

print(is_palindrome("Never odd or even"))
print(is_palindrome("I am samit poudel"))
```

Example 5. Calculate the factorials of a number using recursion

```python
##recursion means inside a function you are calling the same function again and agian 
def factorial(num):
    if num==0:
        return 1
    else:
        return num * factorial(num-1)
    
print(factorial(5))
```

example 6: a function to read a file and count the frequency of each word



```python
def count_word_frequency(file_path):
    word_count={}
    with open(file_path, "r") as file:
        for line in file:
            words=line.split()
            for word in words:
                word=word.lower().strip(" '.,!?:;/ ")
                word_count[word]=word_count.get(word,0)+1



    return word_count

file_path="sample.txt"
word_frequency=count_word_frequency(file_path)
print (word_frequency)
```
