```python
##declaring multiple variables 
##you cannot start a variable name with a number or 
##  special character except underscore(_)
##always start variable name with a letter or underscore(_)
age=25
name="LQQ"
height=5.9
print("Age:", age)
print("Name:", name)
print("Height:", height) 
```

```python
##variable name are case sensitive
Age=30
age=25
print("Age:", Age)
print("Age:", age)
```

```python
##type checking and conversion
age=22
print(type(age))
##type conversion
age_str=str(age)
print(age_str)
print(type(age_str))

```

```python
##stribng into integer
num_str="100"
print(type(num_str))
num_int=int(num_str)
print(num_int)
print(type(num_int))
```

```python
##but you cannot convert non-numeric strings to integers
##this will raise a ValueError
##example
name="LQQ"
#int(name)  # This line will raise an error  
```

```python
##floating number integer conversion
height=5.9
print(type(height))
height_int=int(height)
print(height_int)
print(type(height_int))
```

```python
##integer to float
age=22
age_float=float(age)
print(age_float)
print(type(age_float))
```

```python
##dynamic typing in python
##python aloows you to change the type of variable
#as the program executes
var=10
print(var, type(var))

var="Hello"
print(var, type(var))

var=5.5
print(var, type(var))
```
