### Iterators
 Iterators are advanced Python concepts that allow efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
my_list=[1,2,3,4,5,6]
##interate in normal manner
for i in my_list:
    print(i)

```

```python
print(my_list)
```

```python
## now using a iterator using advance method

iterator=iter(my_list)
print(type(iterator))
##this is a class of list iterator
```

```python
print(iterator)
## i cannot see all the elements of the list like i use to see in the above normal method
## what i am able to see is a class of iterator at a certain memory location
```

```python
## in order to iterate and display the elements of the list we will use next()

print(next(iterator))
## this is know as lazy loading as we load 1 value at a time with out exposing the other values of the list
## but after the last elemet of the list is displayed it cannot display any other elements as the are no next eements and it will show error
```

```python
list_1=[1,2,3,4,5]

iterator_1=iter(list_1)
```

```python
## in order to handel that error

try:
    print(next(iterator_1))
except StopIteration:
    print("you have accessed the last element already cant access more")
```

```python
## for string values
str="hello"
str_iterator=iter(str)
```

```python
try:
    print(next(str_iterator))
except StopIteration:
    print("you have accessed the last alphabet already cant access more")
```
