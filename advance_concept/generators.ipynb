### Generators 
Generators are a simple way to create iterators. They use the yield keyword to produce a series of values lazily , which means they generate values on the fly and do not storing them or loading them in the memory.

yeilds simply returns multiple values

```python
def square_1(n):
    for i in range(n):
        return i**2
    
```

```python
square_1(3)

"""
Why this happens

return immediately exits the function

The loop starts with i = 0

So the function returns 0**2 = 0 on the very first iteration

The loop never continues

That's why you always get only one value, and it's 0.


"""
```

```python
## using yeild
def square(n):
    for i in range(n):
        yield i**2

## it iterates and return the square of all the values

"""
What this does

yield returns one value at a time

The function does not stop after the first value

It remembers its state and continues from where it left off

"""
```

```python
square(4)## so now this becomes a generator object containing all the squares from 0 to 3

for i in square(4):
    print(i)
## this will print all the numbers from gnertor object
```

```python
## to display the values
for value in square(5):
    print(value)
```

```python
## an0ther way to call all the iterated elements
a=square(4)## this a is a generator object


```

```python
## to iterate generator object a i can simpy do next

print(next(a))
## here also there will be a stop iteration error
```

### Practical Example: Reading Large files
Generators are particulary usefulfor reading large files because they allow you to process one line at a time without loading the entire file into memory

```python
### practical : reading large files

def read_large_files(file_path):
    with open(file_path,"r") as file:
        for line in file:
            yield line
```

```python
file_path="my_file.txt"

for line in read_large_files(file_path):
    print(line.strip())

### In this way, I can display the lines of the file without storing them in the memory.

"""
line.strip() removes unwanted whitespace from a string.

What exactly does it remove?

It removes whitespace characters from both ends (start and end) of the line:

Spaces " "

Tabs \t

Newlines \n

Carriage returns \r
"""
```
