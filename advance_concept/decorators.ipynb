### Decorators. 
Decorators are powerful and flexible feature in the Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of the decorators including how to create and use them.


```python
##function copy
def welcome():
    return "welcome to advance python practise"

welcome()
```

```python
## copying the welcome function
wel=welcome
print(wel())##calling the cpied function
del welcome##deleting the original function
print(wel())##calling the cpied function after deleting the original function

## her we can see that even thogh i have deleted the original function but the copied function which contains the content of the original function will not be deleted
```

```python
##closures
## its simply function inside another function

def main_welcome(msg1):##providiing variable at the runtime
    msg2="welcome again"
    def sub_welcome_method():
        print(msg1)
        print(msg2)
        print("this is a advance python practise")
        print("i am always learning")
    return sub_welcome_method()

main_welcome("hello")


## here we can see that enclosed functions also can access the variables of the outer functions  
## anything that is defined outside can also be accessed inside
```

```python
## this is a another method here insted of giving a variable we can also give a function that will call the inbuilt function


def main_welcome(func):
    msg2="welcome again"
    def sub_welcome_method():
        func("hi my name is samit")

    return sub_welcome_method()


```

```python
## decorator
def main_welcome(func):
    msg2="welcome again"
    def sub_welcome_method():
        
        print("welcome to advance python practise")
        func()
        print("i am always learning")

    return sub_welcome_method()
```

```python
## now lets make a new function
def course_introductin():
    print("course intoduction -this is advance python concepts")

```

```python
## now i can call this function course_intofuction() inside the main_welcome(func):
main_welcome(course_introductin)##the func() inside main_welcome(func) will call the functiom course_introduction()
```

```python
## by using decorators i can do this by 
@main_welcome
def course_introductin():
    print("course intoduction -this is advance python concepts")
```

```python
###another decorator example
def main_welcome(func):
    msg2="welcome again"
    def wrapper():
        
        print("this is before calling this function")
        func()
        print("this is after calling the function")

    return wrapper  ## not using () as we have to call the below function
```

```python
@main_welcome
def say_hello():
    print("hello")
```
