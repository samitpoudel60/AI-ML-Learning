### File operation, read and write files. 
File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. These lessons will cover the basics of file handling, including reading, writing text files and binary files.

```python
##read a whole file

with open("example.txt","r") as file:
    content=file.read()## .read() will read the entire contents of file
    print(content)
```

```python
##read file line by line
with open("example.txt","r") as file:
    for line in file:
        print(line.strip())## strip removes the new line character 
```

```python
##writing in a file
with open("example.txt","w") as file:
    file.write("hello world \n")
    file.write("this is a new line")

##this ovewrites the previous content of the file and removes it for new sentences
##so this is a bad practise
```

```python
##write a file without overwriting
## by using append mode (a)
with open("example.txt","a") as file:
    file.write("append operation taking place \n")
    file.write("this is a new line \n")

##this does not ovcerlaps the previous stored data and info
    
```

```python
##writing list of lines multiple lines in a file
lines=["first line \n","second line \n","third line \n","fourth line \n"]
with open("example.txt","a") as file:
    file.writelines(lines)
```

Binary files are files that store data in binary (0s and 1s) format, not as human-readable text. They can contain any kind of dataâ€”numbers, images, audio, compiled code, sensor data, etc

```python
##binary files
## .bin extension files are binary files

##write operation in the binary file
data=b"\x00\x01\x02\x03\x04"
with open("example.bin","wb") as file:## wb means write binary
    file.write(data)
```

```python
##read operation in the binary file
with open("example.bin","rb") as file:## rb means read binary
    content=file.read()
    print(content)
```

```python
##read the content from source text file and write to a destination text file
#copying a text file
with open("example.txt","r") as source_file:## wb means write binary
    content=source_file.read()

with open("destination.txt","w") as dest_file:
    dest_file.write(content)
```
