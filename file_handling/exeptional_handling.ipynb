### Understanding Exceptions
 Exception handling in Python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program. This lesson will cover the basics of exceptions, including how to use try, except, else,  and finally blocks.

### What are exceptions? 
Exceptions are events that disrupt the normal flow of the program. They are the abnormal conditions. They occur when an error is encountered during the program execution. Common exceptions include 

ZeroDivisionError -divided by zero

FileNotFoundError-file not found 

ValueError - invalid value

TypeError- invalid type

```python
try:
    result=1/4
    a=b
except ZeroDivisionError as zeroerror:
    print(zeroerror)
    print("please enter a denominator greater than zero")
except Exception as ex1:
    print(ex1)
    print("main exception caught here")
    
```

except Exception is used as a final safety net â€” it catches any other exception that you did not specifically handle above.

```python
try:
    num=int(input("enter a number"))
    result=10/num
except ValueError:
    print("please enter a valid number")
except ZeroDivisionError:
    print("please enter a denominator greater than zero")
except Exception as ex:
    print(ex)
```

The else block is executed if there are no exceptions. If there are no exceptions then the else block has to be executed. If there are exceptions then the exception block will be executed and else block will not be executed. But if there are no exceptions the else block must be executed.


```python
## try except and else block
try:
    num=int(input("enter a number"))
    result=10/num
except ValueError:
    print("please enter a valid number")
except ZeroDivisionError:
    print("please enter a denominator greater than zero")
except Exception as ex:
    print(ex)
else:
    print("the result is ", result)
```

Finally, block is executed. Even if there is an error and exception block is executed, or even if there is no error and else block is executed, the finally block will be executed. So, it doesn't matter whether we run exception or not, the finally block will always be executed. Remember that.

```python
## try except,  else block and finally

try:
    num=int(input("enter a number"))
    result=10/num
except ValueError:
    print("please enter a valid number")
except ZeroDivisionError:
    print("please enter a denominator greater than zero")
except Exception as ex:
    print(ex)
else:
    print("the result is ", result)
finally:
    print("execution complete")
```

```python
## some practical examples
## file handeling and exeprtion handling


try:
    file=open("example1.txt","r")## another way of opening a file
    content=file.read()
    print(content)
except FileNotFoundError:
    print("file not found")
finally:
    if "file" in locals() and  not file.closed:##"file" in locals() Checks if a variable named "file" exists in your code
        file.close()
        print("file closed")

```
