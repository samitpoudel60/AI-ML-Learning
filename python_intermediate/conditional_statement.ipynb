```python
##if statement\
age = 20
if age>=18:
    print("You can vote")
```

```python
##else statement
age = 16
if age>=18:
    print("You can vote")
else:
    print("You cannot vote")
```

```python
##else if statement ((elif))
age = 16
if age<=6:
    print("You are a child")
elif age<=19:
    print("You are teenager")
elif age>=65:
    print("You are senior citizen")
else:
    print("You are an adult")
```

```python
##nested conditional statement
## you can place one or more if, elif, else statements inside another if, elif or else statement.

##Example number even odd or negative

num = int(input("Enter a number: "))

if num>0:
    print("the number is positive")
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
elif num==0:
    print("the number is zero")
else:
    print("the number is negative")
```

```python
##practical example
##determine wheter the year is leap year or not
year = int(input("Enter a year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year," is a leap year")
        else:
            print(year," is not a leap year")
    else:
        print(year," is a leap year")
else:
    print(year," is not a leap year")
```

```python
##calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator=input("Enter opertaion you want to perform (+,-,*,/): ")
if operator=="+":
    print(num1,"+",num2,"=",num1+num2)
elif operator=="-":
    print(num1,"-",num2,"=",num1-num2)  
elif operator=="*":
    print(num1,"*",num2,"=",num1*num2)
elif operator=="/":
    if num2!=0:
        print(num1,"/",num2,"=",num1/num2)
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Invalid operator, please enter a valid operation")
```
