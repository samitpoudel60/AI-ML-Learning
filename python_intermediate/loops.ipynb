```python
for i in range(5): ##0 to 4 will be printed
                ##last number of range 5 is not executed as we start from 0                            
    print(i)
```

```python
for i in range(0,5):#same ho
    print(i)
```

```python
for i in range(0,10,2):#0 to 10 with step size of 2 ((2 value increment hunxa))
    print(i)
```

```python
for i in range(5,1,-1):## -1 means decrement by 1((this way we can decrement))
    print(i)
```

```python
name = "samit poudel"
for i in name:
    print(i)
```

```python
##while loop 
## the while loop will run until the condition is true
count=0
while count<5:
    print(count)
    count=count+1
```

```python
count=0
while count%2==0:
    print(count,"this is even number")
    count=count+1   
```

```python
##loop control statements
#break statement
##beak statement is used to terminate the loop

for i in range(10):
    if i==7:
        break##7 aayesi loop terminate hunxa
    print(i)
```

```python
##continue statement
##it skips the current iteration and continues with the next iteration

for i in range(10):
    if i==5:
        continue##5 aayesi print hudaina ani agadi badxa
    print(i)##5 print vaxaina 5 print hunna but loop continues
```

```python
##pass
##it is a null operation NOP and it does nothing
##
for i in range(10):
    if i==5:
        print("this is pass statement")
        pass##5 aayesi kehi hudaina ani agadi badxa
    print(i)
```

```python
##nested loops
##loop inside another loop

for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
```

```python
##example sum of n natural numbers
sum=0
num=int(input("enter a number"))
for i in range(num+1):
    sum=i+sum
print("the sum is",sum)
```

```python
##while loop
sum=0
num=int(input("enter a number"))
while num>0:
    sum=sum+num
    num=num-1
print("the sum is",sum)
```
