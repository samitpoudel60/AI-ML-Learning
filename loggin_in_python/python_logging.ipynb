PYTHON LOGGING

Logging is a crucial aspect of any application, providing a way to track events, errors, and operational information. Python built-in logging modules offer a flexible framework for emitting log messages from Python programs. In this lesson, we'll cover the basics of logging, including how to configure logging, log levels, and best practices for using logging in Python application

Python logging is a built-in system in Python used to record messages about what your program is doing while it runs.
It helps you debug, track errors, and monitor program behavior without using lots of print() statements.

Yes üëç Python logging is similar to console.log() in JavaScript, but it is much more powerful.

 Why logging is important

Instead of printing messages everywhere, logging lets you:

‚úÖ See errors and warnings

‚úÖ Track program flow and events

‚úÖ Save messages to a file

‚úÖ Control how much detail you want (debug vs production)

‚úÖ Turn logs ON/OFF without changing code

Logging levels (very important)

Python has 5 standard log levels (from lowest to highest severity):

Level	Use case

DEBUG---	Detailed info for debugging typically of intrest only while diagnosing problem

INFO----	General program info( confirmation that things are working as expected)

WARNING---	Something unexpected but not fatal idication that something unexpected happen but not fatal so the software is still working as expected

ERROR----	 due to more Serious problem, the software has not been able to perform

CRITICAL---very serious problem indicating that the program Program may crash program may be unable to continue to running 

```python
import logging
##we need to configure basic logging settingsbefore importing and using it
logging.basicConfig(level=logging.DEBUG) ##there are five levels of logging: DEBUG, INFO, WARNING, ERROR, CRITICAL

##log messages with different severity levels
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")


```

```python
##also we can configure various things in our logging settings inorder to display 
## to display more detailed log messages
import logging
logging.basicConfig(   ##always remember to restart the kernel because above we already have configured logging and one=ce we have configured it we cannot reconfigure it in the same session
    level=logging.DEBUG,
    filename='app.log',  ##this will save the log messages to a file named app.log
    filemode='w',  ##this will overwrite the log messages to the file instead of appending if we had to add the log message we would use append but we have to overwrite so we are using 'w'
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S' ,##this is for the date time format 
    force=True


)
##the name is displaying root because we are using the root logger here if we had created a custom logger using logging.getLogger('custom_logger') then the name would have been custom_logger


##log messages with different severity levels
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```
