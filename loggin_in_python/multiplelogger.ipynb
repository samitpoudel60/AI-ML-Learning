```python
##now we will learn to create custom_named logger instad of root logger 
##so basically we will create logger for two different modules and we will see how to create custom named logger for each module

import logging
logging.basicConfig(
    level=logging.DEBUG,  ##this will set the logging level to debug so that all the messages with level debug and above will be logged
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', ##this will save the log messages to a file named
    datefmt='%Y-%m-%d %H:%M:%S' ,##this is for the date time format 
    force=True
)


##now lests make alogger for module1
logger1 = logging.getLogger('module1')  ##this will create a custom named logger for module1 named 'module1'
logger1.setLevel(logging.DEBUG)  ##setting the logging level for module1 logger this will overwrite the basicConfig level if different

##now lets make a logger for module2
logger2 = logging.getLogger('module2')  ##this will create a custom named logger for module2 named 'module2'
logger2.setLevel(logging.ERROR)  ##setting the logging level for module2 logger this will overwrite the basicConfig level if different



```

```python
##log message with different loggers 
logger1.debug("This is a debug message from module1")
logger2.warning("This is a warning message from module2")## it will not be  shown because module2 logger level is set to ERROR it will only show error and above level messages
logger2.critical("This is an error message from module2")
logger1.error("This is an error message from module1")
```
